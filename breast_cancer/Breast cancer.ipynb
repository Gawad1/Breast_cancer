```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
raw_data = pd.read_csv('Cancer.csv')
```

```python
raw_data.head()
```

```python
raw_data.info()
```

```python
#drop id beacuse it is irrelevant
raw_data=raw_data.drop('id',axis=1)
```

```python
#encoding M/B
diagnosis= pd.get_dummies(raw_data['diagnosis'])
data = pd.concat([diagnosis,raw_data], axis=1)
data = data.drop('diagnosis',axis=1)
data = data.drop('B',axis=1)
data.rename(columns={"M": "malignant"}, inplace=True)
data
```

```python
#corrolation heatmap to help selecting the features
corr = data.corr().round(2)

f, ax = plt.subplots(figsize=(20, 20))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, cmap=cmap, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.tight_layout()
```

```python
#drop the worst columns because they are highly collorated with the means
worst_cols = ['radius_worst', 
        'texture_worst', 
        'perimeter_worst', 
        'area_worst', 
        'smoothness_worst', 
        'compactness_worst', 
        'concavity_worst',
        'concave points_worst', 
        'symmetry_worst', 
        'fractal_dimension_worst']
data=data.drop(worst_cols,axis=1)
```

```python
#drop cols related to area and perimeter and keeping the radius
cols = ['perimeter_mean',
        'perimeter_se', 
        'area_mean', 
        'area_se']
data=data.drop(cols,axis=1)
```

```python
data.columns
```

```python
#new corrolation heatmap 
corr = data.corr().round(2)

f, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corr, cmap=cmap, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
plt.tight_layout()
```

```python
X=data.drop('malignant',axis=1)
y=data['malignant']
```

```python
#splitting data to train and validation 
X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.3, random_state=40)
```

```python
s = StandardScaler()
X_train=s.fit_transform(X_train)
X_val=s.transform(X_val)
```

```python
logistic_model=LogisticRegression()
logistic_model.fit(X_train,y_train.values.ravel())
pred=logistic_model.predict(X_val)
```

```python
cm = confusion_matrix(y_val,pred)
cm
```

```python
cm = confusion_matrix(y_val,pred,normalize='true')
cm
```

```python
print(classification_report(y_val,pred))
```

```python
print(accuracy_score(y_val,pred))
```
